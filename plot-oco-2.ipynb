### Plot OCO-2 XCO2

```python
import xarray as xr
import rioxarray
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
oco2_l2 = xr.open_dataset('oco2_LtCO2_211231_B10206Ar_220228193805s.nc4')
```

```python
oco2_l2
```

```python
plt.figure(figsize=(15,7))
plt.scatter(oco2_l2.longitude, oco2_l2.latitude, s=50, c=oco2_l2.xco2, cmap='magma')
plt.colorbar(label='xco2')
```
